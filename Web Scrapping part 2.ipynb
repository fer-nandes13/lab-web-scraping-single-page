```python
url = "https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

table_rows = soup.select('#mw-content-text table.wikitable tbody tr')
data = []
for row in table_rows[:11]:  # Select only the first 10 rows
    language_name = row.select_one('td:nth-child(1) a')
    speakers = row.select_one('td:nth-child(2)')
    language_family = row.select_one('td:nth-child(3) a')
    branch = row.select_one('td:nth-child(4) a')
    if language_name and speakers and speakers.text.strip().replace('.', '').isdigit() and language_family and branch:
        data.append({
            "Language": language_name.text.strip(),
            "Speakers": float(speakers.text.strip().replace(',', '')),
            "Language Family": language_family.text.strip(),
            "Branch": branch.text.strip()
        })

df = pd.DataFrame(data)
df
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the target Wikipedia page
url = "https://en.wikipedia.org/wiki/Python"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the table with the specified summary table's caption
    summary_table = soup.select_one('#mw-content-text > div.mw-content-ltr.mw-parser-output > table.wikitable > tbody')
    
    if summary_table:
        # Find all the links within the table
        links = summary_table.find_all('a', href=True, title=True)
        
        # Extract the href and title attributes of the links
        link_data = [{'title': link['title'], 'href': link['href']} for link in links]
        
        # Print the list of links and titles
        for item in link_data:
            print("Title:", item['title'])
            print("Href:", item['href'])
            print()
    else:
        print("Summary table not found on the page.")
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)

```

```python
import requests
from bs4 import BeautifulSoup

# URL of the target Wikipedia page
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all the links on the page
    links = soup.find_all('a', href=True)
    
    # Extract the href attributes of the links
    link_data = [link['href'] for link in links]
    
    # Print the list of links
    for item in link_data:
        print("Link:", item)
else:
    print("Failed to retrieve the webpage. Status code:", response.status_code)
```
