```python
import pandas as pd
```

```python
data_audio_features = pd.read_csv("df_w_audio_ft_j_ver.csv")
```

```python
data_audio_features
```

```python
# K-means
import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler
from matplotlib.lines import Line2D
```

```python
from sklearn.cluster import KMeans
```

```python
num_audio_features = data_audio_features.select_dtypes(include = np.number)
```

```python
num_audio_features
```

```python
X_audio = StandardScaler().fit_transform(num_audio_features)
pd.DataFrame(X_audio,columns = num_audio_features.columns).head()
```

```python
kmeans = KMeans(n_clusters=3, random_state=1234)
kmeans.fit(X_audio)
```

```python
# Predicting / assigning the clusters:
clusters = kmeans.predict(X_audio)

# Check the size of the clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
clusters
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(X_audio[:, 0], X_audio[:, 3], c=clusters,alpha=0.5)
plt.show()
```

```python
kmeans.inertia_
```

```python
# Choosing k

import numpy as np
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_audio)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
from sklearn.metrics import silhouette_score
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_audio)
    silhouette.append(silhouette_score(X_audio, kmeans.predict(X_audio)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Slhouette score showing the optimal k')
```

```python
kmeans = KMeans(n_clusters=18, random_state=1234).fit(X_audio)
clusters_1 = kmeans.predict(X_audio)
pd.Series(clusters_1).value_counts().sort_index()
```
